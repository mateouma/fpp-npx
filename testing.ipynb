```python
template_amps = np.max(np.abs(templates), axis=1) # maximum amplitude of each template on each channel
template_amps.shape
```

```python
sort_template_to_clust = {t:c for t,c in zip(spk_templates, spk_clusters)}
```

```python
chan_temps = []
templates_on_chan_list = []

amp_thres = 0

for i,chan in enumerate(chan_map):
    templates_on_chan = np.where(template_amps[:,i] > amp_thres)[0]
    templates_on_chan_list.append(templates_on_chan)
    chan_temps.append(len(templates_on_chan))

    print('Channel', chan)
    print(templates_on_chan) # template indices that are located on channel i
```

```python
units_inuse = np.array(wf['cluId'], dtype='int')
units_inuse
```

```python
chan_template_dict = {}
chan_unit_dict_uiu = {} # the units we're using

for wgc in wf_gu_channels:
    chan_idx = np.where(chan_map == wgc)[0][0]

    sorted_templates = [sort_template_to_clust.get(tmp) for tmp in templates_on_chan_list[chan_idx]]

    chan_template_dict[f'ch{wgc}'] = templates_on_chan_list[chan_idx]
    chan_unit_dict_uiu[f'ch{wgc}'] = np.intersect1d(units_inuse, sorted_templates)
```
